For unzipping zipped folders on google drive (faster than using the zip extractor)

```python
import zipfile
from google.colab import drive
drive.mount('/content/drive')
```

```python
zip_file_path = r'/content/drive/My Drive/Cropped Stacks Test/06.zip' #unzipping each participant
extraction_dir = '/content/drive/My Drive/Cropped Stacks Test'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_dir)
```
